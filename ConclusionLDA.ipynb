1) **¿Cuál sería la mejor?**

**¿Por qué?**  
- LDA tiene el **recall más alto** (0.545), es decir, detecta más de la mitad de los enfermos reales.
- NaiveBayes tiene el **F1 más alto** (0.332) y un recall también muy bueno (0.417), mostrando buen balance.
- Ambos modelos tienen **buen ajuste** sin sobreajustarse.

---

2) **¿Por qué la has elegido?**

En un contexto de salud, priorizamos **encontrar a la mayor cantidad posible de enfermos**, lo que se refleja en el recall. LDA detecta el 54.5% de los enfermos reales, mientras que NaiveBayes detecta el 41.7%. Aunque ambos tienen precisión baja (se equivocan diagnosticando como enfermos a algunos sanos), esto es preferible a dejar enfermos sin diagnosticar. El F1 score elevado en ambos casos confirma un buen equilibrio general.

---

3) **¿Qué la diferencia del resto aparte de las métricas?**

- **LDA y NaiveBayes** priorizan encontrar enfermos (recall alto), mientras modelos como GradientBoosting, StackingClassifier o LGBM priorizan precisión o accuracy general.
- LDA es un modelo lineal que busca la mejor proyección de los datos para separar clases.
- NaiveBayes utiliza probabilidades condicionales asumiendo independencia entre variables.
- Ambos son más simples que modelos como XGBoost o RandomForest, lo que los hace más interpretables.
- Los modelos más complejos tienen mejor accuracy general pero dejan escapar más enfermos, lo que en salud representa un riesgo inaceptable.

---

4) **Explica cada métrica con ejemplos para niños (caso salud):**

- **Accuracy:**  
  "De cada 100 niños, el doctor adivina correctamente si 81 (LDA) u 86 (NaiveBayes) están sanos o enfermos."
- **Precision:**  
  "Si el doctor dice que 10 niños están enfermos, solo 2-3 realmente lo están (a veces asusta a niños sanos)."
- **Recall:**  
  "Si hay 10 niños enfermos, el doctor LDA encuentra a 5-6, y el doctor NaiveBayes encuentra a 4-5 de ellos."
- **F1:**  
  "Es la nota final que mezcla si el doctor encuentra enfermos y si no asusta a los sanos."
- **Diferencia abs:**  
  "Si en clase y en el examen final adivinas igual de bien, aprendiste de verdad y no solo memorizaste."

---

5) **¿Qué método matemático usa LDA y en qué se diferencia?**

- **LDA** (Linear Discriminant Analysis) busca una combinación lineal de variables que mejor separa los grupos (enfermos vs. sanos). Matemáticamente, calcula proyecciones lineales que maximizan la distancia entre las medias de las clases y minimizan la dispersión dentro de cada clase.

- **Diferencia con otros modelos:**
  - **NaiveBayes:** Aplica el teorema de Bayes asumiendo independencia entre variables, calculando probabilidades para cada clase.
  - **Gradient Boosting, XGBoost, LGBM:** Son modelos secuenciales de árboles que aprenden de los errores de árboles previos.
  - **RandomForest:** Combina muchos árboles independientes entrenados con diferentes muestras.
  - **LogisticRegression:** También es lineal, pero modela directamente la probabilidad usando la función logística.

- **Simplificando**: LDA es como trazar una línea que mejor separa a los niños sanos de los enfermos considerando sus características, mientras que NaiveBayes es como calcular probabilidades de estar enfermo basado en cada síntoma por separado.

## Explicación de métricas para niños

### Accuracy - "El poder de adivinar bien"
Imagina que eres un doctor mágico que tiene que adivinar si los niños tienen gripe o no.
- **NaiveBayes:** De cada 100 niños, adivinas correctamente si 86 están enfermos o sanos.
- **LDA:** De cada 100 niños, adivinas correctamente si 81 están enfermos o sanos.


```python
import matplotlib.pyplot as plt

# Datos para el gráfico de Accuracy
modelos = ['NaiveBayes', 'LDA']
accuracy = [0.865, 0.808]

plt.figure(figsize=(8, 5))
plt.bar(modelos, accuracy, color=['skyblue', 'lightgreen'])
plt.ylim(0, 1)
plt.title('Poder de adivinar de los doctores mágicos')
plt.ylabel('Accuracy (¿Cuánto acierta?)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(accuracy):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center')

# Añadir imágenes para niños
plt.figtext(0.22, 0.6, "🧙‍♂️", fontsize=25)  # NaiveBayes
plt.figtext(0.72, 0.55, "👨‍⚕️", fontsize=25)  # LDA

plt.tight_layout()
plt.show()
```

### Precision - "No asustar a los que están sanos"
Si dices "¡Tú tienes gripe!", ¿cuántas veces aciertas?
- **NaiveBayes:** De cada 10 niños que dices "tienen gripe", solo 2-3 realmente la tienen. ¡Asustas a varios por error!
- **LDA:** De cada 10 niños que dices "tienen gripe", solo 2 realmente la tienen. También asustas a varios.

### Recall - "Encontrar a TODOS los enfermos"
Si hay niños con gripe escondidos en el patio, ¿cuántos encuentras?
- **NaiveBayes:** Si hay 10 niños con gripe, encuentras a 4 de ellos.
- **LDA:** Si hay 10 niños con gripe, encuentras a 5-6. ¡Encuentras más enfermos!

```python
# Datos para el gráfico de Recall
modelos = ['NaiveBayes', 'LDA']
recall = [0.417, 0.545]

plt.figure(figsize=(8, 5))
plt.bar(modelos, recall, color=['#FF9999', '#99FF99'])
plt.ylim(0, 1)
plt.title('¿Cuántos niños enfermos encuentras?')
plt.ylabel('Recall (Capacidad de encontrar enfermos)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(recall):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center')

# Añadir iconos para representar enfermos encontrados
plt.figtext(0.25, 0.3, "🔍👧👦👧👦", fontsize=15)  # NaiveBayes encuentra algunos
plt.figtext(0.73, 0.4, "🔍👧👦👧👦👧👦", fontsize=15)  # LDA encuentra más

plt.tight_layout()
plt.show()
```

### F1 - "La nota final"
Es como tu calificación final combinando si encuentras a los enfermos y no asustas a los sanos.
- **NaiveBayes:** Tu nota es 3.3 sobre 10. Eres bastante bueno encontrando enfermos, pero asustas a varios niños sanos.
- **LDA:** Tu nota es 3.1 sobre 10. Eres mejor encontrando enfermos, pero también asustas a varios niños sanos.

### Diferencia abs - "¿Memorizas o aprendes de verdad?"
- **NaiveBayes:** Diferencia 0.000 - Eres igual de bueno con los niños que conoces y con los nuevos. ¡Aprendiste de verdad!
- **LDA:** Diferencia 0.024 - Eres casi igual de bueno con los niños nuevos. No solo memorizaste, también aprendiste.

## Ejemplo narrativo para un niño

### El juego de los doctores mágicos

Imagina que estamos jugando a los doctores mágicos. Hay 100 niños en el patio, y algunos tienen gripe escondida que no se ve fácilmente.

**Doctor NaiveBayes:**
- De 100 niños, adivina correctamente si 86 están enfermos o sanos.
- Si hay 10 niños con gripe escondida, encuentra a 4 de ellos.
- Cuando dice "tú tienes gripe", solo acierta 2-3 veces de cada 10. Asusta a varios niños sanos.
- Su nota final es 3.3 sobre 10, porque encuentra algunos enfermos y no asusta demasiado.

**Doctor LDA:**
- De 100 niños, adivina correctamente si 81 están enfermos o sanos.
- Si hay 10 niños con gripe escondida, encuentra a 5-6. ¡Encuentra más enfermos!
- Cuando dice "tú tienes gripe", solo acierta 2 veces de cada 10. También asustas a varios niños sanos.
- Su nota final es 3.1 sobre 10, porque encuentra más enfermos pero asusta a muchos sanos.

**¿Cuál elegirías?**
- Si lo más importante es no dejar escapar a ningún niño enfermo, elige al Doctor LDA.
- Si quieres un doctor que no se equivoque tanto en general, elige al Doctor NaiveBayes.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Simulamos resultados para explicar a un niño
# Real: [0=Sano, 1=Enfermo]
# Predicción: [0=Sano, 1=Enfermo]

# Para NaiveBayes
y_true = np.concatenate([np.zeros(85), np.ones(15)])  # 85 sanos, 15 enfermos
# NaiveBayes encuentra 6 de 15 enfermos (recall ~0.42) y tiene precision ~0.28
y_pred_nb = np.concatenate([np.zeros(70), np.ones(15), np.zeros(9), np.ones(6)])

# Para LDA
# LDA encuentra 8 de 15 enfermos (recall ~0.54) y tiene precision ~0.22
y_pred_lda = np.concatenate([np.zeros(57), np.ones(28), np.zeros(7), np.ones(8)])

# Matriz de confusión para NaiveBayes
cm_nb = confusion_matrix(y_true, y_pred_nb)
disp_nb = ConfusionMatrixDisplay(confusion_matrix=cm_nb, 
                               display_labels=["Sano", "Enfermo"])

# Matriz de confusión para LDA
cm_lda = confusion_matrix(y_true, y_pred_lda)
disp_lda = ConfusionMatrixDisplay(confusion_matrix=cm_lda,
                                display_labels=["Sano", "Enfermo"])

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Dibujar las matrices de confusión
disp_nb.plot(ax=ax1, values_format="d", colorbar=False)
disp_lda.plot(ax=ax2, values_format="d", colorbar=False)

# Titulos
ax1.set_title("Doctor NaiveBayes")
ax2.set_title("Doctor LDA")

# Añadir emojis para niños
ax1.text(-0.2, -0.2, "🧙‍♂️", fontsize=24, transform=ax1.transAxes)
ax2.text(-0.2, -0.2, "👨‍⚕️", fontsize=24, transform=ax2.transAxes)

# Explicación para niños
fig.suptitle("¿Cómo adivina cada doctor?", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# Gráfico para niños sobre los tipos de errores
labels = ['Doctor NaiveBayes', 'Doctor LDA']
error_general = [1-0.865, 1-0.808]  # Tasa de error general
error_grave = [1-0.417, 1-0.545]    # No detectar enfermos

x = np.arange(len(labels))  # Ubicación de las etiquetas
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, error_general, width, label='Errores generales', color='orange')
rects2 = ax.bar(x + width/2, error_grave, width, label='¡No encontrar enfermos!', color='red')

# Añadir etiquetas, título y leyenda
ax.set_ylabel('% de errores')
ax.set_title('¿Cuánto se equivoca cada doctor?')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Función para etiquetar las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1%}',
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Añadir iconos de caritas para representar los errores
plt.figtext(0.25, 0.15, "🙂", fontsize=24)  # NaiveBayes - pocos errores generales
plt.figtext(0.75, 0.2, "😕", fontsize=24)  # LDA - más errores generales
plt.figtext(0.35, 0.35, "😟", fontsize=24)  # NaiveBayes - más errores graves
plt.figtext(0.85, 0.25, "😊", fontsize=24)  # LDA - menos errores graves

fig.tight_layout()
plt.show()
```

## Conclusión para un niño

Imagina que tienes que elegir entre dos doctores mágicos:

1. **Doctor NaiveBayes**: Es como un doctor que no siempre encuentra a todos los enfermos, pero se equivoca menos en general. De cada 10 niños enfermos, encuentra a 4, y cuando dice que alguien está enfermo, acierta 2-3 veces de cada 10.

2. **Doctor LDA**: Es como un doctor más preocupado que encuentra más niños enfermos, pero también se equivoca más. De cada 10 niños enfermos, encuentra a 5-6, pero cuando dice que alguien está enfermo, solo acierta 2 veces de cada 10.

**¿A cuál elegirías?** Depende de lo que te dé más miedo:
- Si te da más miedo que un niño enfermo no reciba tratamiento, elige al Doctor LDA.
- Si prefieres un doctor que casi siempre tenga razón, elige al Doctor NaiveBayes.

En medicina, muchas veces preferimos al Doctor LDA, porque es mejor encontrar a más niños enfermos, aunque a veces nos equivoquemos y asustemos a algunos niños sanos.

```python
# Diagrama comparativo: "LAS DOS REDES PARA PECES ENFERMOS"
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(14, 7))
plt.axis('off')

# --- Parámetros de las redes ---
# NaiveBayes (izquierda)
nb_x, nb_y = 1.5, 4
nb_width, nb_height = 3, 4
# LDA (derecha)
lda_x, lda_y = 7.5, 4
lda_width, lda_height = 3, 4

# --- Dibujar redes ---
def draw_net(ax, x, y, width, height, nx, ny, color, lw=1.5, alpha=0.7):
    # Líneas verticales
    for i in range(nx+1):
        ax.plot([x + i*width/nx]*2, [y, y+height], color=color, lw=lw, alpha=alpha)
    # Líneas horizontales
    for j in range(ny+1):
        ax.plot([x, x+width], [y + j*height/ny]*2, color=color, lw=lw, alpha=alpha)

# Red NaiveBayes: fina (más líneas)
draw_net(ax, nb_x, nb_y, nb_width, nb_height, nx=8, ny=8, color='gray', lw=2)
# Red LDA: amplia (menos líneas)
draw_net(ax, lda_x, lda_y, lda_width, lda_height, nx=4, ny=4, color='gray', lw=2)

# --- Peces enfermos (rojos) y sanos (verdes) ---
# NaiveBayes: 4 enfermos atrapados, 6 fuera; 1 sano atrapado
nb_enfermos_atrapados = [(2, 6), (2.5, 5), (3, 7), (1.8, 4.5)]
nb_enfermos_fuera = [(0.7, 3.5), (1.2, 3.2), (2.8, 3.1), (3.2, 4.2), (2.2, 3.7), (3.5, 5.5)]
nb_sano_atrapado = [(2.7, 4.2)]

# LDA: 6 enfermos atrapados, 4 fuera; 4 sanos atrapados
lda_enfermos_atrapados = [(8, 7), (8.5, 6.5), (9, 7.5), (7.8, 5.5), (9.2, 6.2), (8.2, 5.2)]
lda_enfermos_fuera = [(7.2, 3.5), (8.7, 3.2), (9.5, 4.2), (7.5, 4.7)]
lda_sanos_atrapados = [(8.7, 5.7), (9.1, 5.5), (8.5, 6.1), (7.9, 6.3)]

# Dibujar peces enfermos atrapados (rojo, corazón roto)
for (x, y) in nb_enfermos_atrapados:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "💔", fontsize=18)
for (x, y) in nb_enfermos_fuera:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', alpha=0.4, zorder=4)
    ax.text(x-0.1, y+0.2, "💔", fontsize=18, alpha=0.4)
for (x, y) in nb_sano_atrapado:
    ax.scatter(x, y, s=400, color='limegreen', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "✅", fontsize=18)

for (x, y) in lda_enfermos_atrapados:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "💔", fontsize=18)
for (x, y) in lda_enfermos_fuera:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', alpha=0.4, zorder=4)
    ax.text(x-0.1, y+0.2, "💔", fontsize=18, alpha=0.4)
for (x, y) in lda_sanos_atrapados:
    ax.scatter(x, y, s=400, color='limegreen', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "✅", fontsize=18)

# --- Carteles métricas ---
# NaiveBayes
ax.text(nb_x+nb_width/2, nb_y+nb_height+0.5, "Red NaiveBayes\nRed fina y selectiva", fontsize=13, ha='center', fontweight='bold')
ax.text(nb_x+nb_width/2, nb_y-0.7, "Recall: 41.7% | Precisión: 27.6%\nDeja escapar 6 enfermos", fontsize=11, ha='center', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# LDA
ax.text(lda_x+lda_width/2, lda_y+lda_height+0.5, "Red LDA\nRed amplia y protectora", fontsize=13, ha='center', fontweight='bold')
ax.text(lda_x+lda_width/2, lda_y-0.7, "Recall: 54.5% | Precisión: 22.1%\nSalva + vidas, aunque tenga + errores", fontsize=11, ha='center', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- Flecha de decisión ---
ax.annotate('', xy=(lda_x+lda_width/2, lda_y+lda_height+1), xytext=(nb_x+nb_width/2, nb_y+nb_height+1),
            arrowprops=dict(facecolor='deepskyblue', edgecolor='deepskyblue', arrowstyle='->', lw=3))
ax.text((nb_x+lda_x+lda_width)/2, nb_y+nb_height+1.3, "Priorizamos salvar vidas:\n¡Mejor detectar 54.5% de enfermos que dejar escapar 58.3%!", 
        fontsize=13, ha='center', color='deepskyblue', fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- Título y leyenda ---
plt.title("¿Qué red eligirías para proteger el estanque?", fontsize=18, fontweight='bold', y=1.08)
ax.text((nb_x+lda_x+lda_width)/2, nb_y-1.5, 
        "En salud, un falso positivo duele menos que un falso negativo. Por eso elegimos LDA.",
        fontsize=13, ha='center', color='dimgray', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- Caja de texto destacada ---
ax.text((nb_x+lda_x+lda_width)/2, nb_y-2.3, 
        "Aunque NaiveBayes tiene mejor F1 (0.332 vs 0.315), LDA detecta +30% de casos graves.",
        fontsize=12, ha='center', color='crimson', fontweight='bold', bbox=dict(facecolor='mistyrose', alpha=0.8, boxstyle='round,pad=0.4'))

# --- Leyenda visual ---
peces_rojos = mpatches.Patch(color='red', label='Pez enfermo (💔)')
peces_verdes = mpatches.Patch(color='limegreen', label='Pez sano (✅)')
ax.legend(handles=[peces_rojos, peces_verdes], loc='lower right', fontsize=12, framealpha=0.9)

plt.tight_layout()
plt.show()
```
