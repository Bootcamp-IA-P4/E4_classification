---

3) **¬øQu√© la diferencia del resto aparte de las m√©tricas?**

- **LDA y NaiveBayes** priorizan encontrar enfermos (recall alto), mientras modelos como GradientBoosting, StackingClassifier o LGBM priorizan precisi√≥n o accuracy general.
- LDA es un modelo lineal que busca la mejor proyecci√≥n de los datos para separar clases.
- NaiveBayes utiliza probabilidades condicionales asumiendo independencia entre variables.
- Ambos son m√°s simples que modelos como XGBoost o RandomForest, lo que los hace m√°s interpretables.
- Los modelos m√°s complejos tienen mejor accuracy general pero dejan escapar m√°s enfermos, lo que en salud representa un riesgo inaceptable.

4) **Explica cada m√©trica con ejemplos para ni√±os (caso salud):**

- **Accuracy:**  
  "De cada 100 ni√±os, el doctor adivina correctamente si 81 (LDA) u 86 (NaiveBayes) est√°n sanos o enfermos."
- **Precision:**  
  "Si el doctor dice que 10 ni√±os est√°n enfermos, solo 2-3 realmente lo est√°n (a veces asusta a ni√±os sanos)."
- **Recall:**  
  "Si hay 10 ni√±os enfermos, el doctor LDA encuentra a 5-6, y el doctor NaiveBayes encuentra a 4-5 de ellos."
- **F1:**  
  "Es la nota final que mezcla si el doctor encuentra enfermos y si no asusta a los sanos."
- **Diferencia abs:**  
  "Si en clase y en el examen final adivinas igual de bien, aprendiste de verdad y no solo memorizaste."

---

5) **¬øQu√© m√©todo matem√°tico usa LDA y en qu√© se diferencia?**

- **LDA** (Linear Discriminant Analysis) busca una combinaci√≥n lineal de variables que mejor separa los grupos (enfermos vs. sanos). Matem√°ticamente, calcula proyecciones lineales que maximizan la distancia entre las medias de las clases y minimizan la dispersi√≥n dentro de cada clase.

- **Diferencia con otros modelos:**
  - **NaiveBayes:** Aplica el teorema de Bayes asumiendo independencia entre variables, calculando probabilidades para cada clase.
  - **Gradient Boosting, XGBoost, LGBM:** Son modelos secuenciales de √°rboles que aprenden de los errores de √°rboles previos.
  - **RandomForest:** Combina muchos √°rboles independientes entrenados con diferentes muestras.
  - **LogisticRegression:** Tambi√©n es lineal, pero modela directamente la probabilidad usando la funci√≥n log√≠stica.

- **Simplificando**: LDA es como trazar una l√≠nea que mejor separa a los ni√±os sanos de los enfermos considerando sus caracter√≠sticas, mientras que NaiveBayes es como calcular probabilidades de estar enfermo basado en cada s√≠ntoma por separado.

## Explicaci√≥n de m√©tricas para ni√±os

### Accuracy - "El poder de adivinar bien"
Imagina que eres un doctor m√°gico que tiene que adivinar si los ni√±os tienen gripe o no.
- **NaiveBayes:** De cada 100 ni√±os, adivinas correctamente si 86 est√°n enfermos o sanos.
- **LDA:** De cada 100 ni√±os, adivinas correctamente si 81 est√°n enfermos o sanos.


```python
import matplotlib.pyplot as plt

# Datos para el gr√°fico de Accuracy
modelos = ['NaiveBayes', 'LDA']
accuracy = [0.865, 0.808]

plt.figure(figsize=(8, 5))
plt.bar(modelos, accuracy, color=['skyblue', 'lightgreen'])
plt.ylim(0, 1)
plt.title('Poder de adivinar de los doctores m√°gicos')
plt.ylabel('Accuracy (¬øCu√°nto acierta?)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(accuracy):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center')

# A√±adir im√°genes para ni√±os
plt.figtext(0.22, 0.6, "üßô‚Äç‚ôÇÔ∏è", fontsize=25)  # NaiveBayes
plt.figtext(0.72, 0.55, "üë®‚Äç‚öïÔ∏è", fontsize=25)  # LDA

plt.tight_layout()
plt.show()
```

### Precision - "No asustar a los que est√°n sanos"
Si dices "¬°T√∫ tienes gripe!", ¬øcu√°ntas veces aciertas?
- **NaiveBayes:** De cada 10 ni√±os que dices "tienen gripe", solo 2-3 realmente la tienen. ¬°Asustas a varios por error!
- **LDA:** De cada 10 ni√±os que dices "tienen gripe", solo 2 realmente la tienen. Tambi√©n asustas a varios.

### Recall - "Encontrar a TODOS los enfermos"
Si hay ni√±os con gripe escondidos en el patio, ¬øcu√°ntos encuentras?
- **NaiveBayes:** Si hay 10 ni√±os con gripe, encuentras a 4 de ellos.
- **LDA:** Si hay 10 ni√±os con gripe, encuentras a 5-6. ¬°Encuentras m√°s enfermos!

```python
# Datos para el gr√°fico de Recall
modelos = ['NaiveBayes', 'LDA']
recall = [0.417, 0.545]

plt.figure(figsize=(8, 5))
plt.bar(modelos, recall, color=['#FF9999', '#99FF99'])
plt.ylim(0, 1)
plt.title('¬øCu√°ntos ni√±os enfermos encuentras?')
plt.ylabel('Recall (Capacidad de encontrar enfermos)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(recall):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center')

# A√±adir iconos para representar enfermos encontrados
plt.figtext(0.25, 0.3, "üîçüëßüë¶üëßüë¶", fontsize=15)  # NaiveBayes encuentra algunos
plt.figtext(0.73, 0.4, "üîçüëßüë¶üëßüë¶üëßüë¶", fontsize=15)  # LDA encuentra m√°s

plt.tight_layout()
plt.show()
```

### F1 - "La nota final"
Es como tu calificaci√≥n final combinando si encuentras a los enfermos y no asustas a los sanos.
- **NaiveBayes:** Tu nota es 3.3 sobre 10. Eres bastante bueno encontrando enfermos, pero asustas a varios ni√±os sanos.
- **LDA:** Tu nota es 3.1 sobre 10. Eres mejor encontrando enfermos, pero tambi√©n asustas a varios ni√±os sanos.

### Diferencia abs - "¬øMemorizas o aprendes de verdad?"
- **NaiveBayes:** Diferencia 0.000 - Eres igual de bueno con los ni√±os que conoces y con los nuevos. ¬°Aprendiste de verdad!
- **LDA:** Diferencia 0.024 - Eres casi igual de bueno con los ni√±os nuevos. No solo memorizaste, tambi√©n aprendiste.

## Ejemplo narrativo para un ni√±o

### El juego de los doctores m√°gicos

Imagina que estamos jugando a los doctores m√°gicos. Hay 100 ni√±os en el patio, y algunos tienen gripe escondida que no se ve f√°cilmente.

**Doctor NaiveBayes:**
- De 100 ni√±os, adivina correctamente si 86 est√°n enfermos o sanos.
- Si hay 10 ni√±os con gripe escondida, encuentra a 4 de ellos.
- Cuando dice "t√∫ tienes gripe", solo acierta 2-3 veces de cada 10. Asusta a varios ni√±os sanos.
- Su nota final es 3.3 sobre 10, porque encuentra algunos enfermos y no asusta demasiado.

**Doctor LDA:**
- De 100 ni√±os, adivina correctamente si 81 est√°n enfermos o sanos.
- Si hay 10 ni√±os con gripe escondida, encuentra a 5-6. ¬°Encuentra m√°s enfermos!
- Cuando dice "t√∫ tienes gripe", solo acierta 2 veces de cada 10. Tambi√©n asustas a varios ni√±os sanos.
- Su nota final es 3.1 sobre 10, porque encuentra m√°s enfermos pero asusta a muchos sanos.

**¬øCu√°l elegir√≠as?**
- Si lo m√°s importante es no dejar escapar a ning√∫n ni√±o enfermo, elige al Doctor LDA.
- Si quieres un doctor que no se equivoque tanto en general, elige al Doctor NaiveBayes.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Simulamos resultados para explicar a un ni√±o
# Real: [0=Sano, 1=Enfermo]
# Predicci√≥n: [0=Sano, 1=Enfermo]

# Para NaiveBayes
y_true = np.concatenate([np.zeros(85), np.ones(15)])  # 85 sanos, 15 enfermos
# NaiveBayes encuentra 6 de 15 enfermos (recall ~0.42) y tiene precision ~0.28
y_pred_nb = np.concatenate([np.zeros(70), np.ones(15), np.zeros(9), np.ones(6)])

# Para LDA
# LDA encuentra 8 de 15 enfermos (recall ~0.54) y tiene precision ~0.22
y_pred_lda = np.concatenate([np.zeros(57), np.ones(28), np.zeros(7), np.ones(8)])

# Matriz de confusi√≥n para NaiveBayes
cm_nb = confusion_matrix(y_true, y_pred_nb)
disp_nb = ConfusionMatrixDisplay(confusion_matrix=cm_nb, 
                               display_labels=["Sano", "Enfermo"])

# Matriz de confusi√≥n para LDA
cm_lda = confusion_matrix(y_true, y_pred_lda)
disp_lda = ConfusionMatrixDisplay(confusion_matrix=cm_lda,
                                display_labels=["Sano", "Enfermo"])

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Dibujar las matrices de confusi√≥n
disp_nb.plot(ax=ax1, values_format="d", colorbar=False)
disp_lda.plot(ax=ax2, values_format="d", colorbar=False)

# Titulos
ax1.set_title("Doctor NaiveBayes")
ax2.set_title("Doctor LDA")

# A√±adir emojis para ni√±os
ax1.text(-0.2, -0.2, "üßô‚Äç‚ôÇÔ∏è", fontsize=24, transform=ax1.transAxes)
ax2.text(-0.2, -0.2, "üë®‚Äç‚öïÔ∏è", fontsize=24, transform=ax2.transAxes)

# Explicaci√≥n para ni√±os
fig.suptitle("¬øC√≥mo adivina cada doctor?", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# Gr√°fico para ni√±os sobre los tipos de errores
labels = ['Doctor NaiveBayes', 'Doctor LDA']
error_general = [1-0.865, 1-0.808]  # Tasa de error general
error_grave = [1-0.417, 1-0.545]    # No detectar enfermos

x = np.arange(len(labels))  # Ubicaci√≥n de las etiquetas
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, error_general, width, label='Errores generales', color='orange')
rects2 = ax.bar(x + width/2, error_grave, width, label='¬°No encontrar enfermos!', color='red')

# A√±adir etiquetas, t√≠tulo y leyenda
ax.set_ylabel('% de errores')
ax.set_title('¬øCu√°nto se equivoca cada doctor?')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Funci√≥n para etiquetar las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1%}',
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# A√±adir iconos de caritas para representar los errores
plt.figtext(0.25, 0.15, "üôÇ", fontsize=24)  # NaiveBayes - pocos errores generales
plt.figtext(0.75, 0.2, "üòï", fontsize=24)  # LDA - m√°s errores generales
plt.figtext(0.35, 0.35, "üòü", fontsize=24)  # NaiveBayes - m√°s errores graves
plt.figtext(0.85, 0.25, "üòä", fontsize=24)  # LDA - menos errores graves

fig.tight_layout()
plt.show()
```

Imagina que tienes que elegir entre dos doctores m√°gicos:

En medicina, muchas veces preferimos al Doctor LDA, porque es mejor encontrar a m√°s ni√±os enfermos, aunque a veces nos equivoquemos y asustemos a algunos ni√±os sanos.

```python
# Diagrama comparativo: "LAS DOS REDES PARA PECES ENFERMOS"
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(14, 7))
plt.axis('off')

# --- Par√°metros de las redes ---
# NaiveBayes (izquierda)
nb_x, nb_y = 1.5, 4
nb_width, nb_height = 3, 4
# LDA (derecha)
lda_x, lda_y = 7.5, 4
lda_width, lda_height = 3, 4

# --- Dibujar redes ---
def draw_net(ax, x, y, width, height, nx, ny, color, lw=1.5, alpha=0.7):
    # L√≠neas verticales
    for i in range(nx+1):
        ax.plot([x + i*width/nx]*2, [y, y+height], color=color, lw=lw, alpha=alpha)
    # L√≠neas horizontales
    for j in range(ny+1):
        ax.plot([x, x+width], [y + j*height/ny]*2, color=color, lw=lw, alpha=alpha)

# Red NaiveBayes: fina (m√°s l√≠neas)
draw_net(ax, nb_x, nb_y, nb_width, nb_height, nx=8, ny=8, color='gray', lw=2)
# Red LDA: amplia (menos l√≠neas)
draw_net(ax, lda_x, lda_y, lda_width, lda_height, nx=4, ny=4, color='gray', lw=2)

# --- Peces enfermos (rojos) y sanos (verdes) ---
# NaiveBayes: 4 enfermos atrapados, 6 fuera; 1 sano atrapado
nb_enfermos_atrapados = [(2, 6), (2.5, 5), (3, 7), (1.8, 4.5)]
nb_enfermos_fuera = [(0.7, 3.5), (1.2, 3.2), (2.8, 3.1), (3.2, 4.2), (2.2, 3.7), (3.5, 5.5)]
nb_sano_atrapado = [(2.7, 4.2)]

# LDA: 6 enfermos atrapados, 4 fuera; 4 sanos atrapados
lda_enfermos_atrapados = [(8, 7), (8.5, 6.5), (9, 7.5), (7.8, 5.5), (9.2, 6.2), (8.2, 5.2)]
lda_enfermos_fuera = [(7.2, 3.5), (8.7, 3.2), (9.5, 4.2), (7.5, 4.7)]
lda_sanos_atrapados = [(8.7, 5.7), (9.1, 5.5), (8.5, 6.1), (7.9, 6.3)]

# Dibujar peces enfermos atrapados (rojo, coraz√≥n roto)
for (x, y) in nb_enfermos_atrapados:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "üíî", fontsize=18)
for (x, y) in nb_enfermos_fuera:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', alpha=0.4, zorder=4)
    ax.text(x-0.1, y+0.2, "üíî", fontsize=18, alpha=0.4)
for (x, y) in nb_sano_atrapado:
    ax.scatter(x, y, s=400, color='limegreen', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "‚úÖ", fontsize=18)

for (x, y) in lda_enfermos_atrapados:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "üíî", fontsize=18)
for (x, y) in lda_enfermos_fuera:
    ax.scatter(x, y, s=400, color='red', marker='>', edgecolor='k', alpha=0.4, zorder=4)
    ax.text(x-0.1, y+0.2, "üíî", fontsize=18, alpha=0.4)
for (x, y) in lda_sanos_atrapados:
    ax.scatter(x, y, s=400, color='limegreen', marker='>', edgecolor='k', zorder=5)
    ax.text(x-0.1, y+0.2, "‚úÖ", fontsize=18)

# --- Carteles m√©tricas ---
# NaiveBayes
ax.text(nb_x+nb_width/2, nb_y+nb_height+0.5, "Red NaiveBayes\nRed fina y selectiva", fontsize=13, ha='center', fontweight='bold')
ax.text(nb_x+nb_width/2, nb_y-0.7, "Recall: 41.7% | Precisi√≥n: 27.6%\nDeja escapar 6 enfermos", fontsize=11, ha='center', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# LDA
ax.text(lda_x+lda_width/2, lda_y+lda_height+0.5, "Red LDA\nRed amplia y protectora", fontsize=13, ha='center', fontweight='bold')
ax.text(lda_x+lda_width/2, lda_y-0.7, "Recall: 54.5% | Precisi√≥n: 22.1%\nSalva + vidas, aunque tenga + errores", fontsize=11, ha='center', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- Flecha de decisi√≥n ---
ax.annotate('', xy=(lda_x+lda_width/2, lda_y+lda_height+1), xytext=(nb_x+nb_width/2, nb_y+nb_height+1),
            arrowprops=dict(facecolor='deepskyblue', edgecolor='deepskyblue', arrowstyle='->', lw=3))
ax.text((nb_x+lda_x+lda_width)/2, nb_y+nb_height+1.3, "Priorizamos salvar vidas:\n¬°Mejor detectar 54.5% de enfermos que dejar escapar 58.3%!", 
        fontsize=13, ha='center', color='deepskyblue', fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- T√≠tulo y leyenda ---
plt.title("¬øQu√© red eligir√≠as para proteger el estanque?", fontsize=18, fontweight='bold', y=1.08)
ax.text((nb_x+lda_x+lda_width)/2, nb_y-1.5, 
        "En salud, un falso positivo duele menos que un falso negativo. Por eso elegimos LDA.",
        fontsize=13, ha='center', color='dimgray', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.3'))

# --- Caja de texto destacada ---
ax.text((nb_x+lda_x+lda_width)/2, nb_y-2.3, 
        "Aunque NaiveBayes tiene mejor F1 (0.332 vs 0.315), LDA detecta +30% de casos graves.",
        fontsize=12, ha='center', color='crimson', fontweight='bold', bbox=dict(facecolor='mistyrose', alpha=0.8, boxstyle='round,pad=0.4'))

# --- Leyenda visual ---
peces_rojos = mpatches.Patch(color='red', label='Pez enfermo (üíî)')
peces_verdes = mpatches.Patch(color='limegreen', label='Pez sano (‚úÖ)')
ax.legend(handles=[peces_rojos, peces_verdes], loc='lower right', fontsize=12, framealpha=0.9)

plt.tight_layout()
plt.show()
```
