### Influencia de variables categóricas sobre la variable objetivo

Se analiza la relación entre cada variable categórica y la variable objetivo usando tablas de contingencia, test chi-cuadrado y gráficos de barras. Esto permite identificar qué variables categóricas tienen mayor asociación con la presencia de enfermedad cardíaca.

```python
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

chi2_results = []
for col in cat_cols:
    if col == 'Heart_Disease':
        continue
    ct = pd.crosstab(df[col], df['Heart_Disease'])
    chi2, p, dof, ex = chi2_contingency(ct, correction=False)
    chi2_results.append({'Variable': col, 'Chi2': chi2, 'p-value': p})

chi2_df = pd.DataFrame(chi2_results).sort_values('p-value')
display(chi2_df)

# Gráfico: top 5 variables categóricas más asociadas (menor p-valor)
top_cat = chi2_df.nsmallest(5, 'p-value')['Variable']
for col in top_cat:
    plt.figure(figsize=(6,3))
    prop = pd.crosstab(df[col], df['Heart_Disease'], normalize='index')
    prop.plot(kind='bar', stacked=True, ax=plt.gca())
    plt.title(f'Distribución de Heart_Disease según {col}')
    plt.ylabel('Proporción')
    plt.xlabel(col)
    plt.legend(title='Heart_Disease')
    plt.tight_layout()
    plt.show()
```

**Explicación:**

- La tabla muestra el estadístico chi2 y el p-valor para cada variable categórica respecto a la variable objetivo.
- Un p-valor bajo (<0.05) indica fuerte asociación entre la variable categórica y la presencia de enfermedad cardíaca.
- Los gráficos muestran cómo se distribuye la variable objetivo dentro de cada categoría, permitiendo identificar patrones claros de riesgo.
Claro. Esa tabla muestra los resultados del **test chi-cuadrado** entre cada variable categórica y la variable objetivo (`Heart_Disease`). Los campos significan:

- **Variable**: Nombre de la variable categórica analizada.
- **Chi2**: Estadístico chi-cuadrado calculado. Un valor alto indica mayor diferencia entre las categorías respecto a la variable objetivo.
- **p-value**: Probabilidad de obtener ese resultado si no hubiera asociación real. Un valor muy bajo (por ejemplo, <0.05) indica que la asociación es estadísticamente significativa.

**Interpretación:**
- Todas las variables listadas tienen un p-value extremadamente bajo (prácticamente 0), lo que significa que **hay una asociación estadísticamente significativa entre cada variable y la presencia de enfermedad cardíaca**.
- Las variables con mayor estadístico Chi2 (por ejemplo, `General_Health`, `Age_Category`, `Diabetes`) son las que muestran mayor diferencia en la distribución de la variable objetivo entre sus categorías, es decir, **son las más asociadas a la presencia o ausencia de enfermedad cardíaca**.

En resumen:  
**Todas esas variables categóricas están fuertemente asociadas con la variable objetivo, y las primeras de la lista son las más relevantes según el test chi-cuadrado.**

```python
# 1. Definir función de validación cruzada con SMOTE y métricas adicionales
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from imblearn.over_sampling import SMOTE

def CrossValidation(X, y, classifier):
    # Realiza validación cruzada estratificada con SMOTE en cada fold
    accuracy = []
    precision = []
    recall = []
    f1 = []
    cv = StratifiedKFold(n_splits=10, random_state=None, shuffle=False)
    for train, test in cv.split(X, y):
        X_train, X_test = X.iloc[train], X.iloc[test]
        y_train, y_test = y.iloc[train], y.iloc[test]
        sm = SMOTE()
        X_train, y_train = sm.fit_resample(X_train, y_train)
        classifier.fit(X_train, y_train)
        prediction = classifier.predict(X_test)
        accuracy.append(metrics.accuracy_score(y_test, prediction))
        precision.append(metrics.precision_score(y_test, prediction, zero_division=0))
        recall.append(metrics.recall_score(y_test, prediction, zero_division=0))
        f1.append(metrics.f1_score(y_test, prediction, zero_division=0))
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

# 2. Calcular importancia de variables y seleccionar las relevantes
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np

X = df_X.dropna()
y = df_y.loc[X.index]

# Entrena GradientBoosting y LDA para obtener importancia de variables
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X, y)
gbc_importances = gbc.feature_importances_

lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
lda_importances = np.abs(lda.coef_[0])
lda_importances = lda_importances / lda_importances.sum()

feat_imp = pd.DataFrame({
    'Variable': X.columns,
    'GradientBoosting': gbc_importances,
    'LDA': lda_importances
})

# 3. Seleccionar variables con importancia >= 0.01 en al menos uno de los modelos
mask = (feat_imp['GradientBoosting'] >= 0.01) | (feat_imp['LDA'] >= 0.01)
vars_to_keep = feat_imp.loc[mask, 'Variable'].tolist()

# 4. Filtrar df_X para solo usar las variables seleccionadas
df_X_reduced = df_X[vars_to_keep]

# Identificar variables numéricas y categóricas seleccionadas
num_cols = ['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption', 'Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']
cat_cols = ['General_Health', 'Checkup', 'Exercise', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Diabetes', 'Arthritis', 'Sex', 'Age_Category', 'Smoking_History']
cat_dummies = [col for col in df_X_reduced.columns if col.startswith('Age_Category_')]
cat_cols_expanded = [col for col in df_X_reduced.columns if col in cat_cols or col in cat_dummies]
num_cols_selected = [col for col in df_X_reduced.columns if col in num_cols]

print(f"Variables seleccionadas para el modelo ({len(vars_to_keep)}): {vars_to_keep}")
print(f"Variables numéricas seleccionadas: {num_cols_selected}")
print(f"Variables categóricas seleccionadas: {cat_cols_expanded}")

# 5. Ajuste de hiperparámetros con GridSearchCV para GradientBoosting
param_grid_gbc = {
    'n_estimators': [30, 50, 100],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 10, 30],
    'min_samples_leaf': [1, 10, 30]
}
gbc_base = GradientBoostingClassifier(random_state=7)
grid_gbc = GridSearchCV(gbc_base, param_grid_gbc, scoring='f1', cv=3, n_jobs=-1, verbose=2)
grid_gbc.fit(df_X_reduced, df_y)
print("Mejores hiperparámetros GradientBoosting:", grid_gbc.best_params_)
print("Mejor F1 score (validación) GradientBoosting:", grid_gbc.best_score_)

# Usa el mejor estimador para entrenar y evaluar con validación cruzada
gbc = grid_gbc.best_estimator_
gbc_cv = CrossValidation(df_X_reduced, df_y, gbc)
gbc_accuracy_mean = np.mean(gbc_cv['accuracy']) * 100
gbc_precision_mean = np.mean(gbc_cv['precision']) * 100
gbc_recall_mean = np.mean(gbc_cv['recall']) * 100
gbc_f1_mean = np.mean(gbc_cv['f1']) * 100

print('Gradient Boosting Classifier (ajustado):')
print(f'  Accuracy:  {gbc_accuracy_mean:.2f}%')
print(f'  Precision: {gbc_precision_mean:.2f}%')
print(f'  Recall:    {gbc_recall_mean:.2f}%')
print(f'  F1 Score:  {gbc_f1_mean:.2f}%')

# 6. Ajuste de hiperparámetros con GridSearchCV para LDA
param_grid_lda = {
    'solver': ['lsqr', 'eigen'],
    'shrinkage': [None, 'auto', 0.1, 0.5, 0.9]
}
lda_base = LinearDiscriminantAnalysis()
grid_lda = GridSearchCV(lda_base, param_grid_lda, scoring='f1', cv=3, n_jobs=-1, verbose=2)
grid_lda.fit(df_X_reduced, df_y)
print("Mejores hiperparámetros LDA:", grid_lda.best_params_)
print("Mejor F1 score (validación) LDA:", grid_lda.best_score_)

# Usa el mejor estimador para entrenar y evaluar con validación cruzada
lda = grid_lda.best_estimator_
lda_cv = CrossValidation(df_X_reduced, df_y, lda)
lda_accuracy_mean = np.mean(lda_cv['accuracy']) * 100
lda_precision_mean = np.mean(lda_cv['precision']) * 100
lda_recall_mean = np.mean(lda_cv['recall']) * 100
lda_f1_mean = np.mean(lda_cv['f1']) * 100

print('Linear Discriminant Analysis (ajustado):')
print(f'  Accuracy:  {lda_accuracy_mean:.2f}%')
print(f'  Precision: {lda_precision_mean:.2f}%')
print(f'  Recall:    {lda_recall_mean:.2f}%')
print(f'  F1 Score:  {lda_f1_mean:.2f}%')
```

### Optimización de hiperparámetros con Optuna para LDA y comparación con GridSearchCV

```python
# --- Optimización de hiperparámetros con Optuna para LDA ---
import optuna
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import StratifiedKFold, cross_val_score

def objective_lda(trial):
    solver = trial.suggest_categorical('solver', ['lsqr', 'eigen'])
    shrinkage = trial.suggest_categorical('shrinkage', [None, 'auto', 0.1, 0.5, 0.9])
    model = LinearDiscriminantAnalysis(solver=solver, shrinkage=shrinkage)
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    score = cross_val_score(model, df_X_reduced, df_y, scoring='f1', cv=cv, n_jobs=-1)
    return score.mean()

study_lda = optuna.create_study(direction='maximize')
study_lda.optimize(objective_lda, n_trials=30)

print("Mejores hiperparámetros LDA (Optuna):", study_lda.best_params)
print("Mejor F1 score (validación, Optuna):", study_lda.best_value)

# Entrena el mejor modelo LDA encontrado por Optuna
best_lda = LinearDiscriminantAnalysis(**study_lda.best_params)
best_lda.fit(df_X_reduced, df_y)

# Validación cruzada con el mejor modelo LDA de Optuna
lda_optuna_cv = CrossValidation(df_X_reduced, df_y, best_lda)
lda_optuna_accuracy_mean = np.mean(lda_optuna_cv['accuracy']) * 100
lda_optuna_precision_mean = np.mean(lda_optuna_cv['precision']) * 100
lda_optuna_recall_mean = np.mean(lda_optuna_cv['recall']) * 100
lda_optuna_f1_mean = np.mean(lda_optuna_cv['f1']) * 100

print('LDA (Optuna):')
print(f'  Accuracy:  {lda_optuna_accuracy_mean:.2f}%')
print(f'  Precision: {lda_optuna_precision_mean:.2f}%')
print(f'  Recall:    {lda_optuna_recall_mean:.2f}%')
print(f'  F1 Score:  {lda_optuna_f1_mean:.2f}%')

# Comparación de resultados GridSearchCV vs Optuna para LDA
import pandas as pd
lda_compare = pd.DataFrame({
    'Métrica': ['Accuracy', 'Precision', 'Recall', 'F1'],
    'GridSearchCV': [lda_accuracy_mean, lda_precision_mean, lda_recall_mean, lda_f1_mean],
    'Optuna': [lda_optuna_accuracy_mean, lda_optuna_precision_mean, lda_optuna_recall_mean, lda_optuna_f1_mean]
})
display(lda_compare)
```

```python
# --- Comparación de LDA Optuna vs GridSearchCV con cuadro de métricas ---

from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.model_selection import cross_val_predict, StratifiedKFold
import pandas as pd
import numpy as np

def get_metrics(model, X, y):
    # Validación cruzada estratificada
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    # Predicciones de validación cruzada (test)
    y_pred_cv = cross_val_predict(model, X, y, cv=cv, n_jobs=-1)
    # Métricas test (CV)
    acc_test = accuracy_score(y, y_pred_cv)
    prec_test = precision_score(y, y_pred_cv, zero_division=0)
    rec_test = recall_score(y, y_pred_cv, zero_division=0)
    f1_test = f1_score(y, y_pred_cv, zero_division=0)
    # Entrenamiento completo (train)
    model.fit(X, y)
    y_pred_train = model.predict(X)
    acc_train = accuracy_score(y, y_pred_train)
    # Diferencia absoluta
    diff = abs(acc_train - acc_test)
    # Tipo de ajuste
    if diff < 0.01 and acc_test > 0.7:
        ajuste = "Buen ajuste"
    elif diff > 0.05:
        ajuste = "Overfitting"
    else:
        ajuste = "Underfitting"
    return {
        "Accuracy Train": acc_train,
        "Accuracy Test": acc_test,
        "Precision Test": prec_test,
        "Recall Test": rec_test,
        "F1 Test": f1_test,
        "Diferencia abs": diff,
        "Tipo de ajuste": ajuste
    }

# Calcula métricas para ambos modelos
metrics_grid = get_metrics(lda, df_X_reduced, df_y)
metrics_optuna = get_metrics(best_lda, df_X_reduced, df_y)

# Construye el DataFrame de resultados
results = pd.DataFrame([
    {
        "Modelo": "LDA GridSearchCV",
        **{k: round(v, 3) if isinstance(v, float) else v for k, v in metrics_grid.items()}
    },
    {
        "Modelo": "LDA Optuna",
        **{k: round(v, 3) if isinstance(v, float) else v for k, v in metrics_optuna.items()}
    }
])

# Reordena columnas
results = results[[
    "Modelo", "Accuracy Train", "Accuracy Test", "Precision Test", "Recall Test", "F1 Test", "Diferencia abs", "Tipo de ajuste"
]]

# Muestra el cuadro
display(results)
```

```python
# Ajuste de umbral de decisión para LDA GridSearchCV y LDA Optuna
# ...existing code...

# --------------------------------------------------------------------------------------
# COMENTARIO EXPLICATIVO SOBRE EL AJUSTE DE UMBRAL DE DECISIÓN
#
# Sí, puedes y deberías usar ese bloque en tu notebook, especialmente en la sección de
# evaluación de modelos después del ajuste de hiperparámetros y antes de comparar métricas finales.
#
# ¿Qué hace y para qué sirve?
# - Ajusta el umbral de decisión: Por defecto, los modelos clasifican como 1 si la probabilidad es >0.5.
#   Aquí puedes cambiar ese umbral (por ejemplo, a 0.3) para que el modelo sea más “sensible”
#   y detecte más positivos (mayor recall), aunque baje la precisión.
# - Evalúa el impacto del nuevo umbral: Calcula accuracy, precision, recall y F1 usando el nuevo umbral,
#   tanto para GradientBoosting como para LDA.
# - Diagnóstico de ajuste: Indica si hay sobreajuste, buen ajuste o falta de ajuste según la diferencia
#   entre accuracy de train y test.
#
# ¿Dónde usarlo?
# Después de entrenar y ajustar tus modelos (gbc, lda) y de hacer el split train/test,
# puedes usar este bloque para ver cómo cambian las métricas al bajar el umbral de decisión.
#
# ¿Qué ajusta?
# - El umbral de decisión (umbral = 0.3 en vez de 0.5).
# - Esto es útil en salud, porque puedes aumentar el recall (detectar más enfermos), aunque baje la precisión.
#
# Resumen:
# Este bloque te permite experimentar con el umbral de decisión y ver si puedes mejorar recall/F1,
# lo cual es clave en problemas médicos. Puedes copiarlo y ejecutarlo después de tu split train/test
# y entrenamiento de modelos.

# Ejemplo para niños: diferencia entre recall y precision
#
# Imagina que tienes una caja con pelotas rojas (enfermos) y de otros colores (sanos).
# Tu tarea es encontrar todas las pelotas rojas.
#
# - Recall (sensibilidad): De todas las pelotas rojas que había, ¿cuántas lograste encontrar?
#   Ejemplo: Hay 10 pelotas rojas. Encuentras 8. Recall = 8/10 = 0.8 (encontraste el 80% de los enfermos).
#
# - Precision (precisión): De todas las pelotas que sacaste diciendo que eran rojas, ¿cuántas realmente eran rojas?
#   Ejemplo: Sacaste 10 pelotas diciendo que eran rojas, pero solo 8 eran rojas y 2 eran de otro color.
#   Precision = 8/10 = 0.8 (el 80% de tus aciertos realmente eran enfermos).
#
# Diferencia clave:
# - Recall: ¿Cuántos enfermos lograste encontrar de todos los que había?
# - Precision: ¿Cuántos de los que dijiste que estaban enfermos realmente lo estaban?
#
# En medicina, a veces es más importante encontrar a todos los enfermos (alto recall), aunque a veces te equivoques y digas que alguien está enfermo cuando no lo está (baja precision).
# --------------------------------------------------------------------------------------

# ...existing code...
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import numpy as np

def get_metrics_with_threshold(model, X, y, threshold=0.3):
    # Validación cruzada estratificada
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    # Predicciones de probabilidad en CV
    y_proba_cv = cross_val_predict(model, X, y, cv=cv, method='predict_proba', n_jobs=-1)[:, 1]
    y_pred_cv = (y_proba_cv > threshold).astype(int)
    # Métricas test (CV)
    acc_test = accuracy_score(y, y_pred_cv)
    prec_test = precision_score(y, y_pred_cv, zero_division=0)
    rec_test = recall_score(y, y_pred_cv, zero_division=0)
    f1_test = f1_score(y, y_pred_cv, zero_division=0)
    # Entrenamiento completo (train)
    model.fit(X, y)
    y_proba_train = model.predict_proba(X)[:, 1]
    y_pred_train = (y_proba_train > threshold).astype(int)
    acc_train = accuracy_score(y, y_pred_train)
    # Diferencia absoluta
    diff = abs(acc_train - acc_test)
    # Tipo de ajuste
    if diff < 0.01 and acc_test > 0.7:
        ajuste = "Buen ajuste"
    elif diff > 0.05:
        ajuste = "Overfitting"
    else:
        ajuste = "Underfitting"
    return {
        "Accuracy Train": acc_train,
        "Accuracy Test": acc_test,
        "Precision Test": prec_test,
        "Recall Test": rec_test,
        "F1 Test": f1_test,
        "Diferencia abs": diff,
        "Tipo de ajuste": ajuste
    }

# Calcula métricas para ambos modelos con umbral ajustado
umbral = 0.3
metrics_grid = get_metrics_with_threshold(lda, df_X_reduced, df_y, threshold=umbral)
metrics_optuna = get_metrics_with_threshold(best_lda, df_X_reduced, df_y, threshold=umbral)

# Construye el DataFrame de resultados
results = pd.DataFrame([
    {
        "Modelo": "LDA GridSearchCV",
        **{k: round(v, 3) if isinstance(v, float) else v for k, v in metrics_grid.items()}
    },
    {
        "Modelo": "LDA Optuna",
        **{k: round(v, 3) if isinstance(v, float) else v for k, v in metrics_optuna.items()}
    }
])

# Reordena columnas
results = results[[
    "Modelo", "Accuracy Train", "Accuracy Test", "Precision Test", "Recall Test", "F1 Test", "Diferencia abs", "Tipo de ajuste"
]]

# Muestra el cuadro
display(results)
```
